```python
import bs4
import pandas as pd
import requests
```

```python
nm=[]
pr=[]
rt=[]




#n=int(input("Enter no. of pages which you want to scrape =   "))
for i in range(1,51):
    url="https://books.toscrape.com/catalogue/page-"+str(i)+".html"
    res=requests.get(url)
    data=bs4.BeautifulSoup(res.text,"html.parser")
    #print(data.prettify())
    
    
    BookPrice=data.find("p",{"class":"price_color"})
    #print(BookPrice)
    BookRating=data.find("article",{"class":"product_pod"})
    BookRating.find('p').get('class')[1]
    BookName=data.find("h3")
    BookName.a.get("title")
    
    
    for j in data.find_all("h3"):
        nm.append(j.a.get("title"))
        
        
    for j in data.find_all("p",{"class":"price_color"}):
        pr.append(j.text.strip()[1:])
        
        
        
    for j in data.find_all("article",{"class":"product_pod"}):
        rt.append(j.find('p').get('class')[1])
        
    data1=pd.DataFrame({"Books Name":nm,"Books Price":pr,"Books Rating":rt})
    #data1
    
    data1.to_excel("Data.xlsx",index=False)
    df=pd.read_excel("Data.xlsx")
    print(df.shape)
    print(df)
    
    i+=1
```
