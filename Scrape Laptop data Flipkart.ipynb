```python
import pandas as pd
import requests
import bs4
```

```python
url="https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY"
```

```python
req=requests.get(url)
```

```python
req
```

```python
soup=bs4.BeautifulSoup(req.text,"html.parser")
soup
print(soup.prettify())
```

```python
Name=soup.find_all("div",{"class":"_4rR01T"})
print(Name)

```

```python
print(len(Name))
```

```python
Price=soup.find_all("div",{"class":"_30jeq3 _1_WHN1"})
print(Price)
```

```python
len(Price)
```

```python
Ratings=soup.find_all("div",{"class":"_3LWZlK"})
print(Ratings)
```

```python
len(Ratings)
```

```python
Description=soup.find_all("div",{"class":"fMghEO"})
print(Description)
```

```python
len(Description)
```

```python
nm=[]
pr=[]
rt=[]
ds=[]
```

```python
for i in Name:
    nm.append(i.text)
    
for i in Price:
    pr.append(i.text)
    
for i in range(len(nm)):
    rt.append(Ratings[i].text)
    
for i in Description:
    ds.append(i.text)
```

```python
data={"Name":nm,"Price":pr,"Ratings":rt,"Description":ds}
df=pd.DataFrame(data)
```
